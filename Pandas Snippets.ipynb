```python
import pandas as pd
```

```python
airports = pd.read_csv("airports.dat")
airports.columns = [
    "Airport ID",
    "Name",
    "City",
    "Country",
    "IATA",
    "ICAO",
    "Latitude",
    "Longitude",
    "Altitude",
    "Time",
    "DST",
    "TZ database time zone",
    "Type",
    "Source"
]
```

```python
""" SELECT
        Airport ID as ID
    FROM
        airports
"""
(airports
     # Alias 
    .rename(columns={"Airport ID": "ID"})
     # One way to SELECT
    [["ID"]]
).head() # cut off 
```

```python
""" SELECT 
        Airport ID,
        City,
        Country
    FROM
        airports
    LIMIT 7
"""
airports[["Airport ID", "City", "Country"]].head(7)
```

```python
""" SELECT 
        Airport ID AS ID,
        City AS port,
        Country AS nationality,
        Altitude AS flightheight
    FROM
        airports
    ORDER BY 
        Altitude
    DESC
    LIMIT 7
"""
(
airports
    # Alias
    .rename(columns={
        "Airport ID": "ID",
        "City"      : "port",
        "Country"   : "nationality",
        "Altitude"  : "flightheight"
    }) 
    # SELECT cols
    .loc[:,[
        "ID", 
        "port", 
        "nationality", 
        "flightheight"
    ]] 
    # ORDER BY flightheight, DESC
    .sort_values("flightheight", ascending=False)
    # LIMIT 7
    .head(7) 
)
```
